```python
# Check unique countries and locations
print("Countries:", df['country'].nunique())
print("Locations:", df['location_name'].nunique())
```

```python
# Normalize temperature
df['temperature_celsius_normalized'] = (
    df['temperature_celsius'] - df['temperature_celsius'].min()
) / (df['temperature_celsius'].max() - df['temperature_celsius'].min())

# Normalize precipitation
df['precip_mm_normalized'] = (
    df['precip_mm'] - df['precip_mm'].min()
) / (df['precip_mm'].max() - df['precip_mm'].min())
```

```python
print(df[['temperature_celsius', 'temperature_celsius_normalized']].head())
print(df[['precip_mm', 'precip_mm_normalized']].head())
```

```python
# Select columns
selected_columns = [
    'country', 'location_name', 'last_updated',
    'temperature_celsius', 'precip_mm', 'humidity',
    'pressure_mb', 'wind_kph', 'gust_kph', 'visibility_km', 'cloud',
    'uv_index', 'feels_like_celsius',
    'air_quality_PM2.5', 'air_quality_PM10',
    'air_quality_Ozone', 'air_quality_Nitrogen_dioxide',
    'air_quality_Sulphur_dioxide', 'air_quality_us-epa-index', 'air_quality_gb-defra-index',
    'wind_direction', 'temperature_fahrenheit', 'precip_in'
]

df_selected = df[selected_columns].copy()
```

```python
# Convert datetime and extract year, month
df_selected['last_updated'] = pd.to_datetime(df_selected['last_updated'])
df_selected['year'] = df_selected['last_updated'].dt.year
df_selected['month'] = df_selected['last_updated'].dt.month
```

```python
# Aggregate monthly by country and location
df_agg = df_selected.groupby(['country', 'location_name', 'year', 'month']).agg({
    'temperature_celsius': 'mean',
    'precip_mm': 'sum',
    'humidity': 'mean',
    'pressure_mb': 'mean',
    'wind_kph': 'mean',
    'gust_kph': 'mean',
    'visibility_km': 'mean',
    'cloud': 'mean',
    'uv_index': 'mean',
    'feels_like_celsius': 'mean',
    'air_quality_PM2.5': 'mean',
    'air_quality_PM10': 'mean',
    'air_quality_Ozone': 'mean',
    'air_quality_Nitrogen_dioxide': 'mean',
    'air_quality_Sulphur_dioxide': 'mean',
    'air_quality_us-epa-index': 'mean',
    'air_quality_gb-defra-index': 'mean',
    'wind_direction': lambda x: x.mode()[0] if not x.mode().empty else np.nan,
    'temperature_fahrenheit': 'mean',
    'precip_in': 'sum'
}).reset_index()
```
